```python
# install dtat from dtat-library
# https://github.jpl.nasa.gov/Europa-PESS/dtat-library

import pandas as pd

import dtat.plot as dtatplot
from dtat.dataconnectors.csvconnector import CSVConnector

import numpy as np
import fake_data
```

```python
# load the data

# DTAT comes bundled with a CSVConnector class which loads CSV files and their data
# into the format that DTAT is expecting.

# DTAT can plot with any data that is in a pandas dataframe with the columns
# scet, value, name
# where scet is a timestamp
# value is the value to plot
# name is the name of the channel or other information
# ert is an optional column header

sample_data = CSVConnector("sampledata.csv")

#we aren't actually going to use this in the rest of these examples in the rest of this notebook, but we kept it in so you know it's here
```

# Param v Time

Plotting a parameter versus time is what you're used to. In this example, we are looking at some (simulated) data of a current and voltage as an instrument is turned on and transitions through init mode into operational.

This clearly isn't the most exciting thing that DTAT does, but we'll use this simple example to show off some of the basics of DTAT.

```python
from copy import copy
fig,c,m,t = dtatplot.make_stacked_graph( # make_stacked_graph is the primary plotting method
    fake_data.instrument_turn_on(),  # this is the data that DTAT plots from
    y_vars = [["Voltage"]]  # this is the y variables to plot
    )  # the x variable to plot by defaults to SCET
fig.show() # the first parameter returned is a plotly GraphObject
```

Notice that DTAT's plotting methods return a handful of objects. The following cells explain what those are.

```python
print (c) # The second parameter returned is the list of unassigned colors
```

```python
print (m) # The third parameter returned is the customization dictionary with options for each trace
```

```python
print (t) # The fourth parameter returned is the list of visible trace variables
```

## DTAT Return Values

The following cell reproduces the last plot, but note that now we are passing in our own custom dict for defining the traces, and our own list of colors to use.

Right now we're using those pretty trivially, but will use them in more sophisticated ways later.

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.instrument_turn_on(), # CSVConnector.get_data() returns formatted data for DTAT
    y_vars = [
        t, # show the same parameters as before on the plot
    ],
    customize_dict=m, # use the same settings for the traces we had before
    unassigned_colors=c # passing the unassigned colors lets DTAT pull from the next color, instead of restarting from red
    )
fig.show() # this should look the same as our previous figure
```

## Plotting more than one parameter agaisnt time.

Let's explore our options for displauing current and voltage together.

Notice that the keyword **y_vars** is a list of lists. This is how you can communicate to DTAT how many subplots you would like to make and how many traces you would like in each subplot.

Each of the lists within the primary **y_vars** list corresponds to a subplot. And each of the list entries in those lists corresponds to a trace on that subplot. Here we only add one trace per subplot, but you can add as many as you want.

You can also add more than two subplots, but note that there is a maxmimum height for a single figure, so you may not be able to add as many as you want.

Notice that the interactive zoom is linked between the two plots.

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.instrument_turn_on(),
    y_vars = [["Voltage"], ["Current"]],  # each sub-list is a plot in the figure
    x_var = "scet"
    )
fig.show()
```

## Plotting multiple traces on one subplot

In this example, we have two traces on the same (and only) subplot.

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.instrument_turn_on(),
    y_vars = [["Voltage","Current"]],  # each sub-list is a plot in the figure
    x_var = "scet"
    )
fig.show()
```

## Setting the height of the graph

Don't like the default heights? Change them!

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.instrument_turn_on(),
    y_vars = [["Voltage"], ["Current"]],
    x_var = "scet",
    figure_height=1000, # the figure width and height can be manually set
    figure_width=1000,
    )
fig.show()
```

## Displaying multiple Traces on the same plot with multiple axes

This probably isn't th best data set to illustrate this point, but if you had two traces with _very_ different Y values, but still want to easily comapre their trends, you can make them each fill up the entire Y axis by adding a second axis with **multi_access=True**.

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.instrument_turn_on(), 
    y_vars = [["Voltage", "Current"]], 
    x_var = "scet", 
    multi_axis=True # multi_axis creates an independent y axis for each y variable
    )
fig.show()
```

## CustomizedTrace

DTAT also offers the CustomizedTrace class to allow you to display your traces however you'd like. The next couple of plots give examples.

```python
from dtat.types import CustomizedTrace # DTAT proviedes a CustomizedTrace class for customization options

fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.instrument_turn_on(), 
    y_vars = [["Voltage", "Current"]], 
    x_var = "scet", 
    multi_axis=True,
    customize_dict={ # by passing the CustomizedTrace in a dictionary with y variables as the keys, you can control more trace settings
        'Voltage': CustomizedTrace(
                                    size=10,  # setting the marker size
                                    symbol='triangle-down-open',  # setting the marker shape, any plotly option is valid
                                    color='skyblue',  # setting the marker color
                                    colorscale='rainbow',  # setting the colorscale of the z axis (see later sections)
                                    showscale=True,  # whether or not to show the colorscale
                                )
                             }
    )
fig.show()
```

```python
from dtat.types import CustomizedTrace

fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.instrument_turn_on(), 
    y_vars = [["Voltage", "Current"]], 
    x_var = "scet", 
    multi_axis=True,
    customize_dict={
        'Voltage': CustomizedTrace(
            size=10,
            #symbol='triangle-down-open',  # options can be omitted to be replaced with the DTAT defaults
            color='skyblue',
            #colorscale='rainbow',
            #showscale=True,
        ),
        'Mode': CustomizedTrace(  # each trace can have its own customization options
            size=7,
            symbol='diamond',
            color='magenta',
        ),
        'not_on_this_plot': CustomizedTrace(  # traces that are not currently visible can be kept with other customization options
            size=10,
            color='yellowgreen',
        )
    }
)
fig.show()
```

# Param v Param

One of the biggest value adds in DTAT is the ability to natively interpolate data for plotting parameters versus parameters.

In this example, we have a different set of simulated current and voltage channels, but instead of plotting them against time, we are plotting them against each other. This section also starts to introduce ways to use the Z axis.

The first plot is a simple plot showing a scatter plot of voltage and current. Note that there seems to be a bit of data on the upper right of the plot that's thinner than the main body of data. I wonder why that is...

```python
param_v_param_plot, c, m, t = dtatplot.make_stacked_graph(
    fake_data.drifting_off_nominal(), 
    y_vars = [["Voltage"]], 
    x_var = "Current", # The x variable doesn't need to be time-based
    plot_lines = False
)

param_v_param_plot.show()

# If there is no direct correlation between the x and y variables,
# DTAT will interpolate each two variable values based on SCET
# using a nearest-value linear interpolation method
```

## Adding the Z axis

By adding **z_var="scet"**, we can direct DTAT to automatically colorize our scatter points based on the scet value of each pair. The more recent redder points seem to explain that data off to the top right. But it's still hard to tell. I bet we can do better...

```python
param_v_param_plot, c, m, t = dtatplot.make_stacked_graph(
    fake_data.drifting_off_nominal(), 
    y_vars = [["Voltage"]], 
    x_var = "Current", # The x variable doesn't need to be time-based
    z_var = "scet",
    plot_lines = False
)

param_v_param_plot.show()

# If there is no direct correlation between the x and y variables,
# DTAT will interpolate each two variable values based on SCET
# using a nearest-value linear interpolation method
```

## Optimizing colors and markers

Here we go! Now it's quite obvious what's going on. We've had recent values go out of family for these points! Better spin up an ART!

```python
from dtat.types import CustomizedTrace

fig,c,m,t = dtatplot.make_stacked_graph(
    fake_data.drifting_off_nominal(), 
    y_vars = [["Voltage"]], 
    x_var = "Current", 
    z_var = "scet",
    plot_lines=False,
    customize_dict={ # Using customize_dict, you can change the colorscale from the default rainbow
        'Voltage': CustomizedTrace(
            size=10,
            symbol="diamond",
            colorscale="viridis",
            showscale=True,
            mode="markers"
            )
        })
fig.show()
```

# Param v Time with binned (discrete) z value coloring

```python
from dtat.types import CustomizedTrace
from dtat.palette import make_discrete_colorscale_from_data
from dtat.datacacher import get_data_from_state

# To apply a binned colorscale, it must be created first

bin_lower_bounds = [40, 50, 60] # the lower values of each bin
# Minimum bounds are inclusive, so the middle bin will start at 50, and the upper bin will start at 60.
# If there are values below the minimum bound, the minimum bound will be extended down to the lowest value
# IE, for this case, 38 will be colored the same color as 40-49.

sample3_data = get_data_from_state(fake_data.instrument_turn_on(), 'Mode') # use DTAT functions to sort out the Sample3 data as our color values

print (sample3_data)

binned_color_scale = make_discrete_colorscale_from_data(
    color_set=['#ff0000', '#ffff00', '#00ff00'], # the set of hex colors to use for each bin, len(color_set) == len(custom_divisions)
    bin_lower_bounds=bin_lower_bounds,
    color_data=sample3_data
)

fig,c,m,t = dtatplot.make_stacked_graph(fake_data.instrument_turn_on(), 
    y_vars = [["Voltage"]], 
    x_var = "scet", 
    z_var = "Mode",
    plot_lines=False,
    customize_dict={ # Using customize_dict, you can apply your new binned color scale
        'Voltage': CustomizedTrace(
            size=10,
            symbol="diamond",
            colorscale=binned_color_scale,
            color='',
            showscale=True,
            mode="markers"
            )
        })
fig.show()
```

# Plotting Events

DTAT also allows you to plot discrete events on plots, and the next two graphs give examples of how that would work.

You can define the events yourself like we have here, but some interesting examples are using a systems-maintained data set in the State Data Store or using Transpire outputs to add in execution times of interesting commands.

Here we;re showing the same thing as above, but in a different way. Arrows point to when the instrument turns on, and when it completes its ramp and autonomously transitions to operational mode. Do instruments actually do that? I don't know. I'm an engineering person. But it's realistic enough for me to make my point...

```python
import datetime

event1 = datetime.datetime.strftime(fake_data.instrument_turn_on()["scet"][300], "%Y-%jT%H:%M:%S.%f")
event2 = datetime.datetime.strftime(fake_data.instrument_turn_on()["scet"][425], "%Y-%jT%H:%M:%S.%f")
event3 = datetime.datetime.strftime(fake_data.instrument_turn_on()["scet"][500], "%Y-%jT%H:%M:%S.%f")

# DTAT allows plotting of events on the figure if time is used as the x variable
'''
events:
        A dictionary
            - key = y_var name
            - value = lists of tuples of (time, name, message)
        example: {"actual": [(time1, "event1", "message1"), (time2, "event2", "message2")], "predicted": [(time3, "event3", "message3")]}
''' 
# the y variable key should be shown on the figure for the events to show up
# the time is when the event should be shown in %Y-%jT%H:%M:%S.%f format
# the name is what displays on the graph and should be short
# the message is shown when the name is hovered over, so it can be longer and more detailed than the name

param_v_time_with_events, c, m, t = dtatplot.make_stacked_graph(
      fake_data.instrument_turn_on(), 
      y_vars = [["Voltage", "Current"]], 
      x_var = "scet", 
      events = {
        "Voltage": [
            (event1, "Instrument Turn on", "The instrument is turned on here, transitioning to init mode and starting to ramp current and voltage"), 
            (event2, "Transition to Operational", "Current/Voltage rame is done and instrument transitions to Operational Mode"), 
        ] 
      }
    )

param_v_time_with_events.show()
```

```python
import datetime

event1 = datetime.datetime.strftime(fake_data.instrument_turn_on()["scet"][300], "%Y-%jT%H:%M:%S.%f")
event2 = datetime.datetime.strftime(fake_data.instrument_turn_on()["scet"][425], "%Y-%jT%H:%M:%S.%f")
event3 = datetime.datetime.strftime(fake_data.instrument_turn_on()["scet"][500], "%Y-%jT%H:%M:%S.%f")

param_v_time_with_events, c, m, t = dtatplot.make_stacked_graph(
      fake_data.instrument_turn_on(), 
      y_vars = [["Voltage", "Current"]], 
      x_var = "scet", 
      events = {
        "Voltage": [
            (event1, "Instrument Turn on", "The instrument is turned on here, transitioning to init mode and starting to ramp current and voltage"), 
            (event2, "Transition to Operational", "Current/Voltage rame is done and instrument transitions to Operational Mode")
        ]
    },
      event_line=True # in addition to the arrow, a black line can be shown across the plot at the time of the event
    )

param_v_time_with_events.show()
```
