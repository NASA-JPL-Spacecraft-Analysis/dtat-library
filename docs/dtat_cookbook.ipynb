```python
import pandas as pd

# install dtat from dtat-library
# https://github.jpl.nasa.gov/Europa-PESS/dtat-library

import dtat.plot as dtatplot
from dtat.dataconnectors.csvconnector import CSVConnector

# location of local mission_data_store home folder; no ending slash
# clone this repo and make sure you are on the `main` branch
# https://github.jpl.nasa.gov/Europa-PESS/mission_data_store
mds_folder = "/Users/winner/mission_data_store"
```

```python
#load the data
predicted = CSVConnector(f"{mds_folder}/data/relay/m20_era/ORB1_LND1_predictlv_feb18-2021_may1-2022.csv")
rlv = CSVConnector(f"{mds_folder}/data/relay/m20_era/ORB1_LND1_rlv_feb18-2021_may1-2022.csv")
```

```python
combined_data = pd.concat([predicted.get_data(), rlv.get_data()], ignore_index=True)
```

# Param v Time

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    predicted.get_data(),  # this is the data that DTAT plots from
    y_vars = [["ORB2_LND1_Predict_RLV"]]  # this is the y variables to plot
    )  # the x variable to plot by defaults to SCET
fig.show()
```

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    combined_data, 
    y_vars = [["ORB2_LND1_Predict_RLV"], ["ORB1_LND1_Actual_RLV"]], 
    x_var = "scet"
    )
fig.show()
```

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    combined_data, 
    y_vars = [["ORB2_LND1_Predict_RLV", "ORB1_LND1_Actual_RLV"]], 
    x_var = "scet",
    figure_height=1000,
    figure_width=1000,
    )
fig.show()
```

```python
fig,c,m,t = dtatplot.make_stacked_graph(
    combined_data, 
    y_vars = [["ORB2_LND1_Predict_RLV", "ORB1_LND1_Actual_RLV"]], 
    x_var = "scet", 
    multi_axis=True
    )
fig.show()
```

```python
from dtat.types import CustomizedTrace

fig,c,m,t = dtatplot.make_stacked_graph(
    combined_data, 
    y_vars = [["ORB2_LND1_Predict_RLV", "ORB1_LND1_Actual_RLV"]], 
    x_var = "scet", 
    multi_axis=True,
    customize_dict={
        'ORB2_LND1_Predict_RLV': CustomizedTrace(
                                    size=10,  # setting the marker size
                                    symbol='triangle-down-open',  # setting the marker shape, any plotly option is valid
                                    color='skyblue',  # setting the marker color
                                    colorscale='rainbow',  # setting the colorscale of the z axis (see later sections)
                                    showscale=True,  # whether or not to show the colorscale
                                )
                             }
    )
fig.show()
```

```python
from dtat.types import CustomizedTrace

fig,c,m,t = dtatplot.make_stacked_graph(
    combined_data, 
    y_vars = [["ORB2_LND1_Predict_RLV", "ORB1_LND1_Actual_RLV"]], 
    x_var = "scet", 
    multi_axis=True,
    customize_dict={
        'ORB2_LND1_Predict_RLV': CustomizedTrace(
                                    size=10,
                                    #symbol='triangle-down-open',  # options can be omitted to be replaced with the DTAT defaults
                                    color='skyblue',
                                    #colorscale='rainbow',
                                    #showscale=True,
                                ),
        'ORB1_LND1_Actual_RLV': CustomizedTrace(  # each trace can have its own customization options
                                    size=7,
                                    symbol='diamond',
                                    color='magenta',
                                ),
        'not_on_this_plot': CustomizedTrace(  # traces that are not currently visible can be kept with other customization options
                                    size=10,
                                    color='yellowgreen',
                                )
                             }
    )
fig.show()
```

# Param v Time for 2 events

```python
import datetime

event1 = datetime.datetime.strftime(combined_data["scet"][100], "%Y-%jT%H:%M:%S.%f")
event2 = datetime.datetime.strftime(combined_data["scet"][300], "%Y-%jT%H:%M:%S.%f")
event3 = datetime.datetime.strftime(combined_data["scet"][500], "%Y-%jT%H:%M:%S.%f")

param_v_time_with_events, c, m, t = dtatplot.make_stacked_graph(combined_data, 
      y_vars = [["ORB2_LND1_Predict_RLV","ORB1_LND1_Actual_RLV"]], 
      x_var = "scet", 
      events = {
        "ORB1_LND1_Actual_RLV": [(event1, "event1", "this is a message for event1")], 
        "ORB2_LND1_Predict_RLV": [(event3, "event3", "message for event 3"), (event2, "event2", "message for event 2")]
      }
    )

param_v_time_with_events.show()
```
